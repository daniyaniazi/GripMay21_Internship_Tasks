```python
imgH , imgW , _ = img.shape 
```

```python
# Displaying Rectangular box arround Character
for boxes in imgBoxes.splitlines():
    boxes= boxes.split(" ")
    x,y,w,h= int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
    cv2.rectangle(img, (x,imgH-y) , (w , imgH-h) , (255,0,0),1)
```

```python
plt.imshow(img) 
```

```python
def DetectText(img):
    imgChar=pytesseract.image_to_string(img)
    imgBoxes = pytesseract.image_to_boxes(img)
    for boxes in imgBoxes.splitlines():
        boxes= boxes.split(" ")
        x,y,w,h= int(boxes[1]),int(boxes[2]),int(boxes[3]),int(boxes[4])
        imgH , imgW , _ = img.shape
        cv2.rectangle(img, (x,imgH-y) , (w , imgH-h) , (255,0,0),2)
    
    print("\n------DETECTED TEXT------\n")
    print(imgChar)
    plt.imshow(img) 
```

```python
img = cv2.imread("test3.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
DetectText(img)
```
